## 1. Download and Explore Toronto Neighborhoods Data

Scrape the Wikipedia page for Neighborhoods in Toronto in order to obtain the data that is in the table of postal codes and transform the data into a pandas dataframe

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
data = pd.read_html(url) 
df=data[0]
```

The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood.
Remove the cells with 'Borough' that is 'Not assigned'

```python
df=df[df.Borough != 'Not assigned']
df
```
